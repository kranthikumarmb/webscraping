```python
# intsalling the libraries 

!pip install bs4
!pip install requests
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#sending requests to page to get access
url='https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'
response=requests.get(url)
soup=BeautifulSoup(response.content,'html.parser')
```

```python
#defining into single lits
movie_name=[]
year=[]
rating=[]
```

```python
movie_data=soup.findAll('div',class_='lister-item mode-advanced')


#using loop to get the whole data
for i in movie_data:
    name=i.h3.a.text
    movie_name.append(name)
    
    realease_year=i.h3.find('span',class_='lister-item-year text-muted unbold').text.replace('(','').replace(')','')
    year.append(realease_year)
    
    rate=i.find('div',class_='inline-block ratings-imdb-rating').text.replace('\n', '')
    rating.append(rate)
```

```python
movie_DF=pd.DataFrame({'movie name':movie_name,'year of release':year,'movie rating':rating})
movie_DF
```

# 3.Write a python program to display all the header tags from wikipedia.org.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as  np
```

```python
url='https://www.imdb.com/list/ls056092300/'
response=requests.get(url)
soup=BeautifulSoup(response.content,'html.parser')
```

```python
movie_name=[]
year_of_release=[]
rating=[]
movie_dat=soup.findAll('div',class_='lister-item mode-detail')
for i in movie_dat:
    name=i.h3.a.text
    movie_name.append(name)
    
    releaseyear=i.h3.find('span',class_='lister-item-year text-muted unbold').text.replace('(', '').replace(')', ' ')
    year_of_release.append(releaseyear)
    
    rate=i.find('div',class_='ipl-rating-star small').text.replace('\n','')
    rating.append(rate)
    
dataframeofmovie=pd.DataFrame({'movie name':movie_name,'released year':year_of_release,'movie rating':rating})
dataframeofmovie
```

# 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

wb=requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(wb.content)
headline=[]
for i in soup.find_all("div",class_="LatestNews-container"):

     headline.append(i.find("a",class_="LatestNews-headline").text)  
time=[]
for i in soup.findAll('time',class_='LatestNews-timestamp'):
    time.append(i.text)
    
newsLink=[]
for i in soup.find_all("div",class_="LatestNews-container"):

     newsLink.append(i.find("a",class_="LatestNews-headline").get("href"))   
    
df=pd.DataFrame({'haedlines':headline,'updated time':time,'newslinks ':newsLink})
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
food=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soap=BeautifulSoup(food.content)
t=[]

for i in soap.find_all('a',class_="restnt-name ellipsis"):
    t.append(i.text)
t    

j=[]

for i in soap.find_all('div',class_='restnt-loc ellipsis'):
    j.append(i.text)
j

s=[]

for i in soap.find_all('span',class_="double-line-ellipsis"):
    s.append(i.text)
s

ima=[]

for i in soap.find_all("img",class_="no-img"):
    ima.append(i.get('data-src'))
ima  

df=pd.DataFrame({'titles':t,'places':j,'images':ima,'prices':s})
df
```
